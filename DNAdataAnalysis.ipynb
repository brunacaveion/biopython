#### Comparison between 2 DNAs

```python
from Bio import SeqIO
from Bio import pairwise2
from Bio.pairwise2 import *
record1=SeqIO.parse('C:/Users/Bruna/OneDrive/Documentos/Biopython/HBB.fasta','fasta')
record2=SeqIO.parse('C:/Users/Bruna/OneDrive/Documentos/Biopython/sequence.fasta','fasta')
HBB=''
bacteria=''
for element in record1:
    HBB=str(element.seq)
for element in record2:
    bacteria=str(element.seq)
```

```python
print(HBB)
```

```python
print(bacteria)
```

```python
globalAlignment=pairwise2.align.globalxx(HBB,bacteria)
globalAlignment # returns a list with all comparison possibilities, it works when both DNAs have same size 
```

```python
format_alignment(*globalAlignment[0]) # alignment with match bars 
```

```python
localAlignment=pairwise2.align.localmx(HBB, bacteria,2,10) # comparisson when the sequences aren't same size
                                                         # 1st number is mathing values, 2nd mismatch
                                # change match and mismatch values doesn't affect the number or nature of comparison
                                # there are localmx, localxx, localxs, etc
```

```python
localAlignment
```

```python
format_alignment(*localAlignment[0]) # check alignment 
```

```python
# Way 1 (Changing Specific element by its index)
bioDict = {'methionine': ['AUG'], 'proline': ['CCC', 'CCTT', 'CCA', 'CCG'], 'stop': ['UAA']}
print('Old Dict', bioDict)
keysList = bioDict.keys()
for var in keysList:
    if var == 'proline':
        for ind, codon in enumerate(bioDict[var]):
            if codon == 'CCTT':
                bioDict[var][ind] = 'CCT'
print('New Dict', bioDict)

# Way 2 (Changing the entire value)
bioDict = {'methionine': ['AUG'], 'proline': ['CCC', 'CCTT', 'CCA', 'CCG'], 'stop': ['UAA']}
print('Old Dict', bioDict)
keysList = bioDict.keys()
for var in keysList:
    if var == 'proline':
        bioDict['proline'] = ['CCC', 'CCT', 'CCA', 'CCG']
print('New Dict', bioDict)
```
